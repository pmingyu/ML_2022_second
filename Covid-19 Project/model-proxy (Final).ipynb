```python
from sklearn.cluster import KMeans, MiniBatchKMeans, AgglomerativeClustering, Birch, SpectralClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# the following is a helper function to visualize the silhouette score
def vslz_silhouette(temp_X, tempN_clusters, temp_cluster_labels, temp_slhtAvg, temp_cluster_centers, algo):
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1
    ax1.set_xlim([-1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(temp_X) + (tempN_clusters + 1) * 10])
        
    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(temp_X, temp_cluster_labels)

    y_lower = 10
    for i in range(tempN_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[temp_cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / tempN_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

        
    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=temp_slhtAvg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(temp_cluster_labels.astype(float) / tempN_clusters)
    ax2.scatter(
        temp_X[:, 0], temp_X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )
    
    if(temp_cluster_centers.shape[0] > 0):
        # Labeling the clusters
        centers = temp_cluster_centers
        # Draw white circles at cluster centers
        ax2.scatter(
            centers[:, 0],
            centers[:, 1],
            marker="o",
            c="white",
            alpha=1,
            s=200,
            edgecolor="k",
        )

        for i, c in enumerate(centers):
            ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(
        "Silhouette analysis for clustering on sample data with n_clusters = %d"
        % tempN_clusters,
        fontsize=14,
        fontweight="bold",
    )
    plt.savefig("%s.jpg"%algo, bbox_inches='tight')
    plt.show()

# a function to calculate and display the silhoueette score and graphs
def silhouette_analysis(dataset, algorithm, msg):
    print(msg)
    
    max_silhouette = -1.1
    min_db = 1000.0
    best_silCluster = 0
    best_dbCluster = 0
    best_cluster_labels = None
    best_model = None
    
    silhouette_score_list = []
    db_score_list = []
    
    # assign dataset from parameter
    X = dataset
    plt.figure(figsize=(10,5))
    for n_clusters in range(3, 21, 1):
        cluster_centers = np.array([])
        # Initialize the clusterer with n_clusters value and a random generator
        # seed of 10 for reproducibility.
        if(algorithm == "kmeans"):
            clusterer = KMeans(n_clusters=n_clusters, random_state=10)
            cluster_labels = clusterer.fit_predict(X)
            cluster_centers = clusterer.cluster_centers_
        elif(algorithm == 'agglomerative'):
            clusterer = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='average')
            cluster_labels = clusterer.fit_predict(X)
        elif(algorithm == 'birch'):
            clusterer = Birch(threshold=0.01, n_clusters=n_clusters)
            cluster_labels = clusterer.fit_predict(X)
        elif(algorithm == 'miniBatchKmeans'):
            clusterer = MiniBatchKMeans(n_clusters=n_clusters)
            cluster_labels = clusterer.fit_predict(X)
            cluster_centers = clusterer.cluster_centers_
        elif(algorithm == 'gaussian'):
            clusterer = GaussianMixture(n_components=n_clusters)
            clusterer.fit(X)
            cluster_labels = clusterer.predict(X)
            cluster_centers = clusterer.means_
        elif(algorithm == 'spectral'):
            clusterer = SpectralClustering(n_clusters=n_clusters)
            cluster_labels = clusterer.fit_predict(X)
            

        # The silhouette_score gives the average value for all the samples.
        # This gives a perspective into the density and separation of the formed
        # clusters
        silhouette_avg = silhouette_score(X, cluster_labels)
        silhouette_score_list.append(silhouette_avg)
        # print("\nFor n_clusters = ", n_clusters, "The average silhouette_score is  = ", silhouette_avg)
        if(silhouette_avg > max_silhouette):
            max_silhouette = silhouette_avg
            best_silCluster = n_clusters
            best_cluster_labels = cluster_labels
            best_model = clusterer
        
        # caculate the db index for each cluster size
        db_score = davies_bouldin_score(X, cluster_labels)
        db_score_list.append(db_score)
        # print("For n_clusters = ", n_clusters, "The DB score is = ", db_score)
        if(db_score < min_db ):
            min_db = db_score
            best_dbCluster = n_clusters
    
    print("Max Silhouette score = ", max_silhouette , " on Cluster = ", best_silCluster)
    print("Min DB score = ", min_db, " on Cluster = ", best_dbCluster)
    
    vslz_silhouette(X, best_silCluster, best_cluster_labels, max_silhouette, cluster_centers, algorithm)
    return (best_model, silhouette_score_list, db_score_list)
```

```python
# asses K-means algorithm using silhouette score
kmeans_best_model, kmeans_silhouette_scores, kmeans_db_scores = silhouette_analysis(patient_cleaned_array, "kmeans", "K-means clustering")
```

```python
# asses Hierarchical agglomerative algorithm using silhouette score
aglmt_best_model, aglmt_silhouette_scores, aglmt_db_scores = silhouette_analysis(patient_cleaned_array, "agglomerative", "Hierarchical Agglomerative clustering")
```

```python
# asses Birch algorithm using silhouette score
birch_best_model, birch_silhouette_scores, birch_db_scores =  silhouette_analysis(patient_cleaned_array, "birch", "Birch clustering")
```

```python
# asses Mini Batch K-means algorithm using silhouette score
miniKmeans_best_model, miniKmeans_silhouette_scores, miniKmeans_db_scores = silhouette_analysis(patient_cleaned_array, "miniBatchKmeans", "Mini batch K-means clustering")
```

```python
# graph model metrics results for the 3 models and compare them
def plot_metrics(kmeans_array, aglmt_array, birch_array, miniKmeans_array, metrics):
    plt.figure(figsize=(10,5))
    plt.title(label=metrics, loc='center')
    plt.ylabel(metrics)
    plt.xlabel('Cluster size')
    plt.xticks([s for s in range(0, 18, 1)], [n for n in range(3, 21, 1)])
    plt.plot(kmeans_array, label='K-means Clustering %s'%metrics)
    plt.plot(aglmt_array, label='Hierarchical Agglomerative Clustering %s'%metrics)
    plt.plot(birch_array, label='Birch Clustering %s'%metrics)
    plt.plot(miniKmeans_array, label='Mini Batch K-means Clustering %s'%metrics)
    plt.legend()
    plt.savefig("model-proxy %s.jpg"%metrics, bbox_inches='tight')
    plt.show()
```

```python
# plot Silhouette score for all models
plot_metrics(kmeans_silhouette_scores, aglmt_silhouette_scores, birch_silhouette_scores, miniKmeans_silhouette_scores, "Silhouette Score")

# plot DB score for all models
plot_metrics(kmeans_db_scores, aglmt_db_scores, birch_db_scores, miniKmeans_db_scores, "DB Score")
```
