```python
# import needed libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
from sklearn import svm
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
# Import Dataset and perform Data pre-processing
def prepare_data():
    sign_data_set = pd.read_csv("../../Datasets/model-sign/Covid Sympotoms Dataset.csv")

    # Drop "Sanitization from Market" column because its less relevant and ~
    # ~ all its values are No
    sign_data_set.drop('Sanitization from Market', axis='columns', inplace=True)

    # Extract Independet variables
    sign_x = sign_data_set.iloc[:, :-1].values

    # Extract the Dependet variable
    sign_y = sign_data_set.iloc[:, -1].values

    # Incode input data (independet variables)
    label_encoders = []
    for i in range(19):
        label_encoder_x = LabelEncoder()
        sign_x[:, i]= label_encoder_x.fit_transform(sign_x[:, i])
        label_encoders.append(label_encoder_x)

    # Incode the ouput data (dependent variable)
    labelencoder_y= LabelEncoder()  
    sign_y = labelencoder_y.fit_transform(sign_y) 

    # split the dataset into traning and validation sets
    sign_Xtrain, sign_Xtest, sign_Ytrain, sign_Ytest= train_test_split(sign_x, sign_y, test_size= 0.15)

    # convert input datasets of traning and testing to float32 for numpy operations
    sign_Xtrain, sign_Xtest = np.array(sign_Xtrain, np.float32), np.array(sign_Xtest, np.float32)
    
    return (sign_Xtrain, sign_Ytrain, sign_Xtest, sign_Ytest)
```

```python
# This section is for functions for training the MLP model

# train the MLP model
def modelMLP_train(xTrain, yTrain, xTest, yTest, itern):
    # determine the number of input features
    n_features = xTrain.shape[1]
    # define model
    model_sign = Sequential()
    model_sign.add(Dense(15, activation='relu', kernel_initializer='he_normal', input_shape=(n_features,)))
    model_sign.add(Dense(25, activation='relu', kernel_initializer='he_normal'))
    model_sign.add(Dropout(0.4))
    model_sign.add(Dense(10, activation='relu', kernel_initializer='he_normal'))
    model_sign.add(Dense(1, activation='sigmoid'))
    # compile the model
    model_sign.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    # configure early stopping (show graph of the impact of early stop)
    early_stop = EarlyStopping(monitor='val_loss', patience=5)
    # fit the model
    model_history = model_sign.fit(xTrain, yTrain, epochs=150, batch_size=32, verbose=0, 
                                   validation_split=0.2, callbacks=[early_stop])
    # evaluate the model
    model_loss, model_acc = model_sign.evaluate(xTest, yTest, verbose=0)
    print("MLP model %s Accuracy = %.4f" % (itern, model_acc))
    return (model_sign, model_history)

# calculate the metrics for the MLP model
def modelMLP_metrics(temp_model, xTest, yTest):
    # predict crisp classes for test set
    yhat_classes = temp_model.predict_classes(xTest, verbose=0)
    # reduce to 1d array
    yhat_classes = yhat_classes[:, 0]
    # accuracy: (tp + tn) / (p + n)
    accuracy = accuracy_score(yTest, yhat_classes)
    # precision tp / (tp + fp)
    precision = precision_score(yTest, yhat_classes)
    # recall: tp / (tp + fn)
    recall = recall_score(yTest, yhat_classes)
    # f1: 2 tp / (2 tp + fp + fn)
    f1 = f1_score(yTest, yhat_classes)    
    return (accuracy, precision, recall, f1)

# train MLP model and return the list of metrics
def MLP():
    # iterate and train the model 25 times and get the average accuracy and best model reference
    mlp_accuracy_list = []
    mlp_precision_list = []
    mlp_recall_list = []
    mlp_f1_list = []

    mlp_max_accuracy = 0
    mlp_best_model = None
    mlp_best_model_history = None
    # iterate and get model and metrics to get better model description
    for itern in range(15):
        # get train and test data
        mlpX_train, mlpY_train, mlpX_test, mlpY_test = prepare_data()
        # train model and get metrics
        iter_model, iter_history = modelMLP_train(mlpX_train, mlpY_train, mlpX_test, mlpY_test, itern)
        iter_acc, iter_prcs, iter_rcl, iter_f1 = modelMLP_metrics(iter_model, mlpX_test, mlpY_test)
        # append results of each iteration to lists
        mlp_accuracy_list.append(iter_acc)
        mlp_precision_list.append(iter_prcs)
        mlp_recall_list.append(iter_rcl)
        mlp_f1_list.append(iter_f1)
        # get the instance and accuracy of the best model
        if(iter_acc > mlp_max_accuracy):
            mlp_max_accuracy = iter_acc
            mlp_best_model = iter_model
            mlp_best_model_history = iter_history
    # calculate the mean of each metrix of the model
    mlp_accuracy_mean = np.mean(mlp_accuracy_list)
    mlp_percision_mean = np.mean(mlp_precision_list)
    mlp_recall_mean = np.mean(mlp_recall_list)
    mlp_f1_mean = np.mean(mlp_f1_list)

    print("\nMean Accuracy = %.4f" % mlp_accuracy_mean)
    print("Mean Percision = %.4f" % mlp_percision_mean)
    print("Mean Recall = %.4f" % mlp_recall_mean)
    print("Mean F1 = %.4f" % mlp_f1_mean)
    return(mlp_accuracy_list, mlp_precision_list, mlp_recall_list, mlp_f1_list, mlp_best_model, 
           mlp_best_model_history)

# call the MLP function
mlp_accuracys, mlp_precisions, mlp_recalls, mlp_f1s, mlp_model, mlp_history = MLP()
```

```python
# This section is for functions for training the SVM model

# trian the SVM model and return the predictions
def modelSVM_train(xTrain, yTrain, xTest, yTest, itern):
    # Use support vector classification
    classifier_sk = svm.SVC()
    # Use the train data to train this classifier
    classifier_sk.fit(xTrain, yTrain)
    # Use the trained model to predict on the test data
    predictions = classifier_sk.predict(xTest)
    score = accuracy_score(yTest, predictions)
    print("SVM model %s Accuracy = %.4f" % (itern, score))
    return (classifier_sk, predictions)

# calculate the model metrix
def modelSVM_metrix(pred, xTest, yTest):
    # accuracy: (tp + tn) / (p + n)
    accuracy = accuracy_score(yTest, pred)
    # precision tp / (tp + fp)
    precision = precision_score(yTest, pred)
    # recall: tp / (tp + fn)
    recall = recall_score(yTest, pred)
    # f1: 2 tp / (2 tp + fp + fn)
    f1 = f1_score(yTest, pred)
    return (accuracy, precision, recall, f1)

# train the SVM model and return the model metrix values for the iterations
def SVM():    
    # iterate and train the model 25 times and get the average accuracy and best model reference
    svm_accuracy_list = []
    svm_precision_list = []
    svm_recall_list = []
    svm_f1_list = []

    svm_max_accuracy = 0
    svm_best_model_prediction = None
    svm_best_model = None

    # iterate and get model and metrics to get better model description
    for itern in range(15):
        # get train and test data
        svmX_train, svmY_train, svmX_test, svmY_test = prepare_data()
        # train model and get metrics
        iter_model, iter_pred = modelSVM_train(svmX_train, svmY_train, svmX_test, svmY_test, itern)
        iter_acc, iter_prcs, iter_rcl, iter_f1 = modelSVM_metrix(iter_pred, svmX_test, svmY_test)

        svm_accuracy_list.append(iter_acc)
        svm_precision_list.append(iter_prcs)
        svm_recall_list.append(iter_rcl)
        svm_f1_list.append(iter_f1)

        if(iter_acc > svm_max_accuracy):
            svm_max_accuracy = iter_acc
            svm_best_model_prediction = iter_pred
            svm_best_model = iter_model

    svm_accuracy_mean = np.mean(svm_accuracy_list)
    svm_percision_mean = np.mean(svm_precision_list)
    svm_recall_mean = np.mean(svm_recall_list)
    svm_f1_mean = np.mean(svm_f1_list)

    print("\nMean Accuracy = %.4f" % svm_accuracy_mean)
    print("Mean Percision = %.4f" % svm_percision_mean)
    print("Mean Recall = %.4f" % svm_recall_mean)
    print("Mean F1 = %.4f" % svm_f1_mean)

    return (svm_accuracy_list, svm_precision_list, svm_recall_list, svm_f1_list, iter_model)

# call the SVM function
svm_accuracys, svm_precisions, svm_recalls, svm_f1s, svm_model = SVM()
```

```python
# This section is for functions for training the LR model

# Design the model-sign based on sklearn Logistic regression model
def modelLR_train(xTrain, yTrain, xTest, yTest, itern):
    # all parameters of the model not specified are set to their defaults
    logisticRegr = LogisticRegression()
    # fit the model to the training data
    logisticRegr.fit(xTrain, yTrain)
    # Use score method to get accuracy of model
    score = logisticRegr.score(xTest, yTest)
    print("LR model %s Accuracy = %.4f" % (itern, score))
    return logisticRegr

# calculate the model metrix
def modelLR_metrix(temp_model, xTest, yTest):
    pred = temp_model.predict(xTest)
    # accuracy: (tp + tn) / (p + n)
    accuracy = accuracy_score(yTest, pred)
    # precision tp / (tp + fp)
    precision = precision_score(yTest, pred)
    # recall: tp / (tp + fn)
    recall = recall_score(yTest, pred)
    # f1: 2 tp / (2 tp + fp + fn)
    f1 = f1_score(yTest, pred)    
    return (accuracy, precision, recall, f1)

# trian the LR model and return the models metrics list
def LR():   
    lr_accuracy_list = []
    lr_precision_list = []
    lr_recall_list = []
    lr_f1_list = []

    lr_max_accuracy = 0
    lr_best_model = None

    # iterate and get model and metrics to get better model description
    for itern in range(15):
        # get train and test data
        lrX_train, lrY_train, lrX_test, lrY_test = prepare_data()
        # train model and get metrics
        iter_model = modelLR_train(lrX_train, lrY_train, lrX_test, lrY_test, itern)
        iter_acc, iter_prcs, iter_rcl, iter_f1 = modelLR_metrix(iter_model, lrX_test, lrY_test)

        lr_accuracy_list.append(iter_acc)
        lr_precision_list.append(iter_prcs)
        lr_recall_list.append(iter_rcl)
        lr_f1_list.append(iter_f1)

        if(iter_acc > lr_max_accuracy):
            lr_max_accuracy = iter_acc
            lr_best_model = iter_model

    lr_accuracy_mean = np.mean(lr_accuracy_list)
    lr_percision_mean = np.mean(lr_precision_list)
    lr_recall_mean = np.mean(lr_recall_list)
    lr_f1_mean = np.mean(lr_f1_list)

    print("\nLR Mean Accuracy = %.4f" % lr_accuracy_mean)
    print("LR Mean Percision = %.4f" % lr_percision_mean)
    print("LR Mean Recall = %.4f" % lr_recall_mean)
    print("LR Mean F1 = %.4f" % lr_f1_mean)
    
    return (lr_accuracy_list, lr_precision_list, lr_recall_list, lr_f1_list, lr_best_model)

# call the LR fucntion
lr_accuracys, lr_precisions, lr_recalls, lr_f1s, lr_model = LR()
```

```python
# graph model metrics results for the 3 models and compare them
def plot_metrics(mlp_array, svm_array, lr_array, metrics):
    plt.figure(figsize=(10,5))
    plt.title(label="Models %s"%metrics, loc='center')
    plt.ylabel(metrics)
    plt.xlabel('Experiment Runs')
    plt.plot(mlp_array, label='MLP %s'%metrics)
    plt.plot(svm_array, label='SVM %s'%metrics)
    plt.plot(lr_array, label='LR %s'%metrics)
    plt.legend()
    plt.savefig("%s.jpg"%metrics, bbox_inches='tight')
    plt.show()
```

```python
# calculate and display the confusion matrix
def cfsn_matrix(testY_data, model_pred, model_type):
    # calculate the confusion matrix
    confusionMatrics = confusion_matrix(testY_data, model_pred)
    # visualize the confusion matrix using matplotlib and seaborn
    plt.figure(figsize=(7,7))
    sns.heatmap(confusionMatrics, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Greens');
    plt.ylabel('Actual label');
    plt.xlabel('Predicted label');
    all_sample_title = '%s model Confusion Matrix'%model_type
    plt.title(all_sample_title, size = 15)
    plt.savefig("%s confusion matrix.jpg"%model_type, bbox_inches='tight')
```
